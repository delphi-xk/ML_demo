```python
import pandas as pd
from pandas import DataFrame
% matplotlib inline
```

```python
traces = pd.read_csv("~/workspace/datasets/trajectories_training.csv")
```

```python
traces.head(10)
```

```python
B_1 = traces.loc[ (traces['intersection_id']=='B') & (traces['tollgate_id']== 1) ].sort_values("starting_time")
```

```python
B_1
```

```python
B_1_time_index = DataFrame(B_1['travel_time'].values,
                           index=pd.to_datetime(B_1['starting_time'], format='%Y-%m-%d %H:%M:%S'),
                           columns=['travel_time'])
```

```python
B_1_resample = B_1_time_index.resample('20Min', closed='left').mean()
```

```python
B_1_resample
```

```python
B_1_resample.plot(figsize=(12,6),title="B to 1 travel time")
```
